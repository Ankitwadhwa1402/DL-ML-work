```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('D:\\Machine learning\\houses_price_citywise.csv')
data
```

# HANDLING THE CATEGORICAL DATA

```python
d1=pd.get_dummies(data['City'],drop_first=True)  #get_dummies will make categorical data into 1 2 3 ,... 
#drop_first will remove the first column to remove dummy variable trap
d1
```

```python
data=pd.concat([data,d1],axis=1)
data
```

```python
data=data.drop('City',axis=1)  #by deafult axis=0
data
```

```python
#model creation

model=LinearRegression()
x=data.iloc[:,[0,1,3,4]]
y=data.iloc[:,[2]]
print(x)
print(y)
```

```python
#splitting the dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
print(x_train)
print(y_train)
```

```python
#train the model

model.fit(x_train,y_train)
pred=model.predict(x_test)
print(pred)  #predicted values
print(y_test)  #real values
```

```python
print(model.intercept_)
print(model.coef_)
```

# EFFICIENCY OF MODEL

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,pred)
print(score)
#43% percent accuracy because of less data
```
