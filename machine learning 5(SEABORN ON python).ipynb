# READ THE DATASET

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips=pd.read_csv('D:\\Machine learning\\tips.csv')
tips.head()
```

```python
import seaborn as sns
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

# LINE PLOT

```python
sns.lineplot(x='size',y='tip',data=tips)
```

## Distribution plot

```python
sns.distplot(tips['tip'])    #which value is more in tip or say frequency of each tip value 
```

### barplot

```python
sns.barplot(x='day',y='tip',data=tips)   #which day tip is maximum
```

```python
#kdeplot

sns.kdeplot(tips['tip'])  #same as distplot but not show bars
```

# regression plot

```python
sns.regplot(x='total_bill',y='tip',data=tips)
```

# Categorical plot

```python
sns.catplot(x='smoker',y='tip',data=tips)
```

```python
#for counting above yes no also we use count plot
sns.countplot(x='smoker',data=tips)
```

## pair plot

```python
#it will pair all the numerical value column and show their plots
sns.pairplot(tips)
```

## Heat map

```python
tips.corr()  #correlation matrix
```

```python
sns.heatmap(tips.corr())  #values are not showing
```

```python
sns.heatmap(tips.corr(),annot=True)
```

```python
sns.heatmap(tips.corr(),annot=True,linewidth=10)
```

```python
sns.heatmap(tips.corr(),annot=True,linewidth=10,square=True) #square shapes
```
