# making the sheets in single excel file


```python
from pandas import ExcelWriter
```

```python
d1=pd.DataFrame({'APPLE':[3,4,0,8,2,5,6],'MANGOES':[5,7,9,3,0,6,7]}) #for sheet 1 in excel
d2=pd.DataFrame({'A':[3,4,5,6],'B':[3,0,6,7]}) #for sheet 2 in excel 2
```

```python
writer=ExcelWriter('excelwithsheets.xlsx')  #created a excel file
```

```python
d1.to_excel(writer,'imp1',index=True)  #d1 stored in writer as a sheet name imp1 
d2.to_excel(writer,'imp2',index=False)    #index is bydefault true if it is false then it will not save the index in sheet
writer.save()  #save the d1 and d2 in excel file
```
