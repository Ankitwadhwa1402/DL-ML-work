### multiple line plot

```python
x=[1,2,3,4,5]
y1=[50,40,70,80,20]
y2=[80,20,20,50,60]
y3=[70,20,60,40,60]
y4=[80,20,20,50,60]

plt.plot(x,y1,label='Enfield',linewidth=1) #linewidth means how much line will be thick
plt.plot(x,y2,label='Honda',linewidth=5)
plt.plot(x,y3,label='Yamaha',linewidth=5)
plt.plot(x,y4,label='Ktm',linewidth=10)
plt.title('Multiple LINE PLOT')
plt.xlabel('Days')
plt.ylabel('Distance in Kms')
plt.legend()   #used to show the label which color has which label
```

```python
x=[1,2,3,4,5]
y1=[50,40,70,80,20]
y2=[80,20,20,50,60]
plt.plot(x,y1,color='green',label='Enfield',linewidth=1) 
plt.plot(x,y2,color='blue',label='Honda',linewidth=5)
plt.grid()  #used to make grids
plt.xlabel('Days')
plt.ylabel('Distance in Kms')
plt.legend()
plt.show()  #remove the output line  <matplotlib.legend.Legend at 0x21912076898>
```

## BAR PLOT

```python
#vertical bar plot  means bars are vertical

x=[10,20,30,40,50]
y=['A','B','C','D','E']
plt.xlabel('quantities')
plt.ylabel('Countries')
plt.title('Sample BAR PLOT')
plt.bar(y,x,color='green')
```

```python
#horizontal bar plot
#only change bar to barh
x=[10,20,30,40,50]
y=['A','B','C','D','E']
plt.xlabel('quantities')
plt.ylabel('Countries')
plt.title('Sample BAR PLOT')
plt.barh(y,x,color='green')  
```

```python
x=[10,20,30,40,50]
y=['A','B','C','D','E']
plt.xlabel('quantities')
plt.ylabel('Countries')
plt.title('Sample BAR PLOT')
plt.bar(y,x,color='green',width=0.2)
```

```python
x=[10,20,30,40,50]
y=['A','B','C','D','E']
plt.xlabel('quantities')
plt.ylabel('Countries')
plt.title('Sample BAR PLOT')
plt.barh(y,x,color='green',height=0.3)
```

### scatter plot

```python
x=[1,2,3,4,5,6,10]
y=[1,4,9,16,25,36,100]
plt.scatter(x,y,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('sample SCATTER PLOT')
plt.show()
```

# pie plot

```python
sub=['DBMS','ML','JAVA','Python','HTML']
students=[23,17,35,29,12]
plt.pie(students,labels=sub,startangle=90)
plt.show()
```

```python
#use of autopct

sub=['DBMS','ML','JAVA','Python','HTML']
students=[23,17,35,29,12]
plt.pie(students,labels=sub,startangle=90,autopct='%1.2f')  # .2f means after decial there will be 2 values
plt.show()
```

```python
#use of startangle

sub=['DBMS','ML','JAVA','Python','HTML']
students=[23,17,35,29,12]
plt.pie(students,labels=sub)
plt.show()
```
