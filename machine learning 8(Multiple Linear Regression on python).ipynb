```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('D:\\Machine learning\\houses_price_3_Columns.csv')
data
```

```python
data.isnull().sum()
```

```python
data.iloc[8,1]=data['Area'].median()
data
```

```python
#independent data(area,rooms) and dependent data(price)
x=data[['rooms','Area']];
y=data[['Price']]
print(x)

print(y)
```

```python
data
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()   #object created


model.fit(x,y)



print(model.intercept_)  #B0
print(model.coef_)  #B1,B2,has 2 value because has 2 features room and area

#new formula is (price=B0+B1*rooms+B2*area)
```

```python
pred=model.predict([[3,250],[4,450]])
pred
```

```python
data1=pd.read_csv("D:\\Machine learning\\houses_input_data.csv")
print(data1)
pred1=model.predict(data1)
pred1
```
