# Support Vector Machine Algorithms used for mainly classification

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("D:\\Machine learning\\Social_Network_Ads.csv")
data
```

```python
#extraction ofID AND DD
x=data.iloc[:,[2,3]].values  #values used to convert into array
y=data.iloc[:,4].values
x
```

```python
y
```

```python
data.isnull().sum()
```

```python
#splitting the data sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)

x_train
```

```python
len(y_train)
```

```python
#features scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()  #object created
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
x_train
```

```python
print(x_test)
```

```python
#create the model(SVM/classification/logistic) whichever is suited

#Fit the SVM model to training dataset
from sklearn.svm import SVC
svm_model=SVC(kernel='rbf')  #bydefault kernel is linear
```

```python
svm_model.fit(x_train,y_train)
```

```python
#prediction
pred=svm_model.predict(x_test)
```

```python
pred
```

```python
#for analysis/accuracy use confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)  #annot for values also visible on colors 
plt.xlabel('actual label')
plt.ylabel('predicted label')
plt.title('confusion matrix')
plt.show()
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,pred)
acc*100

#for more accuracy we change kernel to "rbf","sigmoid" and etc
```
